```python
import pandas as pd

# Assume 'combined_df' is your DataFrame
def find_duplicate_columns(df):
    duplicate_columns = []
    duplicates = set()  # Set to hold all duplicates
    for i in range(df.shape[1]):
        col1 = df.iloc[:, i]
        if col1.name in duplicates:
            continue  # Skip checking if already identified as a duplicate
        for j in range(i + 1, df.shape[1]):
            col2 = df.iloc[:, j]
            if col1.equals(col2):
                duplicate_columns.append((df.columns[i], df.columns[j]))
                duplicates.add(df.columns[j])
                duplicates.add(df.columns[i])
    return duplicate_columns, duplicates

# Check for duplicate columns and get all duplicates
duplicates, all_duplicates = find_duplicate_columns(combined_df)
print("Duplicate Columns:", duplicates)
print("All Duplicate Columns to be Removed:", all_duplicates)

# Remove all duplicate columns from the DataFrame
combined_df_cleaned = combined_df.drop(columns=all_duplicates)

# Print the first few rows of the updated DataFrame
print("Updated DataFrame Preview:")
print(combined_df_cleaned.head())
```
